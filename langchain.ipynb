# Simple chain

```python
from langchain.chat_models import ChatOpenAI
from langchain import PromptTemplate
from langchain.chains import LLMChain

llm=ChatOpenAI(model_name='gpt-3.5-turbo',temperature=0.5)

template=''' You are experienced in virology. explain about the following {virus} in {language}'''

prompt=PromptTemplate(
input_variables=['virus','language'],
template=template
)

chain=LLMChain(llm=llm,prompt=prompt)
output=chain.run({'virus':'nipa','language':'english'})
output
```

# Sequential chains

```python
from langchain.chat_models import ChatOpenAI
from langchain.llms import OpenAI
from langchain import PromptTemplate
from langchain.chains import LLMChain,SimpleSequentialChain

llm1=OpenAI(model_name='text-davinci-003',temperature=0.7,max_tokens=1024)
prompt1= PromptTemplate(
    input_variables=['concept'],
    template=''' You are an experienced Data scientist and python programmer . 
    write a function to code for the {concept}'''
)
chain1=LLMChain(llm=llm1,prompt=prompt1)

llm2=ChatOpenAI(model_name='gpt-3.5-turbo',temperature=1.2)
prompt2=PromptTemplate(
    input_variables=['function'],
    template=''' Give the python function {function} . Explain {function} in detail '''
)

chain2=LLMChain(llm=llm2,prompt=prompt2)

overall_chain=SimpleSequentialChain(chains=[chain1,chain2],verbose=True)

output=overall_chain.run('decision tree')
```

# Splitting and embedding text using Langchain

```python
import os
from dotenv import load_dotenv, find_dotenv
load_dotenv(find_dotenv(), override=True)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
with open('./churchill_speech.txt') as f:
    speech=f.read()


text_splitter=RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=20, # max overlap between chunks needed to maintain some continuity between them
    length_function=len # indicates how length of chunks is calculated (default is count number of characters , but LLM'S use tokens)
    
)
```

```python
chunks=text_splitter.create_documents([speech])
```

```python
chunks[10].page_content
```

```python
print(f'Total chunks {len(chunks)}')
```

# Embedding cost

```python
def print_embedding_cost(texts):
    import tiktoken
    enc = tiktoken.encoding_for_model('text-embedding-ada-002')
    total_tokens = sum([len(enc.encode(page.page_content)) for page in texts])
    print(f'Total Tokens: {total_tokens}')
    print(f'Embedding Cost in USD: {total_tokens / 1000 * 0.0004:.6f}')
    
print_embedding_cost(chunks)
```

# Inserting embeddings into a pinecone index

```python
import os
import pinecone
from langchain.vectorstores import Pinecone

pinecone.init(api_key=os.environ.get('PINECONE_API_KEY'),environment=os.environ.get('PINECONE_ENV'))

```

```python
# deleting all pinecone indexes

indexes=pinecone.list_indexes()
for i in indexes:
    pinecone.delete_index(i)
```

```python
index_name='churchill-speech'
if index_name not in pinecone.list_indexes():
    print(f'creating index {index_name}')
    pinecone.create_index(index_name,dimension=1536,metric='cosine')
    print('completed')
```

# Upload vectors to pinecone using langchain

```python
vector_store=Pinecone.from_documents(chunks,embeddings,index_name=index_name)
```

# Asking questions - Similarity check

```python
query='where should we fight ? '
result=vector_store.similarity_search(query)
print(result)
```

```python
for r in result:
    print(r.page_content)
    print('-'*50)
```

Eventhough the above chunks represent the answer , we cannot give them to the user like this and hence we will be using LLM 

```python
from langchain.chains import RetrievalQA
from langchain.chat_models import ChatOpenAI
```

```python
llm=ChatOpenAI(model='gpt-3.5-turbo',temperature=1)

retriever = vector_store.as_retriever(search_type='similarity',search_kwargs={'k':3})# return 3mostsimilarchunk from user's query

# create chain to answer questions

chain=RetrievalQA.from_chain_type(llm=llm,chain_type="stuff",retriever=retriever) # stuff uses all of the text in the document in the prompt
```

```python
q='where should we fight ?'
answer=chain.run(q)
```
